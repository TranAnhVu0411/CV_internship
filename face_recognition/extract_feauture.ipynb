# Feature Extraction các ảnh mẫu

```python
# Ta sẽ lấy vector đặc trưng sử dụng facenet
def get_feature(image_path, save_path):
    image = cv2.imread(image_path)
    image_tensor = torch.tensor(image)
    image_tensor = image_tensor.reshape((image_tensor.shape[2], image_tensor.shape[0], image_tensor.shape[1])).float()
    img_embedding = resnet(image_tensor.unsqueeze(0))
    img_embedding = img_embedding.detach().numpy()
    np.save(os.path.join(save_path, os.path.basename(image_path)[:-4]+".npy"), img_embedding)
```

```python
# Path setting
source_path = "/Users/trananhvu/Documents/CV/data/preprocess"
save_parent_path = "/Users/trananhvu/Documents/CV/data/feature"
```

```python
for i in os.listdir(source_path):
    if i==".DS_Store":
        continue
    print(i)
    image_source_path = os.path.join(source_path, i)
    for j in os.listdir(image_source_path):
        if j==".DS_Store":
            continue
        print("Image: "+j)
        if not os.path.exists(os.path.join(save_parent_path, i)):
            os.mkdir(os.path.join(save_parent_path, i))
        get_feature(os.path.join(image_source_path, j), os.path.join(save_parent_path, i))
```
